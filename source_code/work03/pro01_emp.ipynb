```python
#Q14) 사원의 이름의 첫 글자가 S로 시작하는 사원의 이름과, 봉급을 출력
# select ename, sal from emp where ename like 'S%';
for res in df['ename']: 
    if res[0]== 'S':
        print(df[['ename','sal']][df['ename']==res])
```

```python
# Q15) 사원의 이름이 T자로 끝나는 사원의 이름과 봉급을 출력
# select ename, sal from emp where ename like '%T';
for res in df['ename']: 
    if res[-1]== 'T':
        print(df[['ename','sal']][df['ename']==res])
```

```python
# Q16) 사원의 이름의 두 번쨰 글자가 M자인 사원의 이름과 , 봉급을 출력
# select ename, sal from emp where ename like '_M%';
for res in df['ename']: 
    if res[1]== 'M':
        print(df[['ename','sal']][df['ename']==res])
```

```python
# 19-1 번외편  : 모평균이 30이고 표준편차가 7인 모집단을 구성
avg = 30
std =7
N=1000000

mogipdan = np.random.randn(N)* std + avg; #np.random.randn(N) 가우시안 정규분포
print(len(mogipdan))
print(mogipdan)
```

```python
# 19-2 번외편 : 위의 19-1의 모집단에서 평균
print(np.mean(mogipdan))
```

```python
#19-3 번외편 : 모평균이 30이고 표준편차가 7인 모집단(1000000)에서 
#표본을 49개를 뽑자 np.choice(대상, 표본개수)
mogipdan = np.random.randn(N)* std + avg;
res = np.random.choice(mogipdan,49)
print(res)
print()
print(f"표본 평균 : {np.mean(res)}")
```
