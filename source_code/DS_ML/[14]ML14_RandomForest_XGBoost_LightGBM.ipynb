### 오차 행렬(confusion Matrix)
* True: 맞음, False: 틀림
* Negative : 0 , Positive : 1

[ [ TN, FP], True Negative, False Positive
<br>
  [ FN, TP] ] False Negative, True Positive
  
정확도 = 맞은갯수/전체갯수 = (TN + TP)/( TN + FP + FN + TP) 

#### 정밀도 = TP/(FP + TP), 예측을  Positive로 한 것 중 Positive답을 맞춘 비율, 양성 예측도
#### 재현율 = TP/(FN + TP), 실제 값이 Positive인 것 중 Positive답을 맞춘 비율, 민감도, TPR(True Positive Rate)
#### F1 스코어  = 2/((1/재현율) + (1/정밀도)) = 2*(재현율*정밀도)/(재현율+정밀도), feature의 중요도, 정밀도와 재현율을 결합한 지표
#### ROC(Receiver Operation Curve) : 수신자 판단 곡선
* TNR(True Negative Rate,특이성) : TN/(FP + TN)
* x축을 FPR(False Positive Rate) : FP/(FP + TN) = 1 - TNR = 1 - 특이성
* y축을 TPR(True Positive Rate,재현율,민감도) : TP/(FN + TP)

####  ROC AUC(Area Under Curve) : ROC 곡선의 면적, 1에 가까울수록 좋음, 1이 최대

```python
### 6. 정확도를 표시할 수 있는 사용자 함수를 만들자
# 정확도 평가(평가 지표)
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.metrics import precision_score, recall_score
from sklearn.metrics import f1_score,roc_auc_score
from sklearn.metrics import classification_report

def get_clf_eval(y_test,pred):
    accuracy = accuracy_score(y_test,pred)
    print('정확도:',accuracy)

    # 오차 행렬
    confusion = confusion_matrix(y_test,pred)
    print('오차행렬:\n',confusion)
    TN = confusion[0,0]
    FP = confusion[0,1]
    FN = confusion[1,0]
    TP = confusion[1,1]

    # 정밀도(precsion) 
    precision = precision_score(y_test,pred)
    print('정밀도:',precision)
    print('TP/(FP + TP):',TP/(FP + TP))
    
    # 재현율(recall)
    recall = recall_score(y_test,pred)
    print('재현율:',recall)
    print('TP/(FN + TP)',TP/(FN + TP))
    
    # f1_score
    f1 = f1_score(y_test,pred)
    print('F1-스코어:',f1)
    print(2/((1/recall) + (1/precision)))
    print(2*(recall*precision)/(recall+precision))
    
    # roc_auc
    roc_auc = roc_auc_score(y_test,pred)
    print('roc_auc:',roc_auc)
```

### Boosting

XGBoost(Kaggle 대회 상위 석권), LightGBM(속도 빠름)
* (1) AdaBoost(Adaptive Boosting) : 개별 약한 학습기에 순차적으로 가중치를 부여해 결합하여 예측
* (2) GBM(Gradient Boost Machine) : 에이다 부스트와 유사하나 가중치를 경사하강법(Gradient Descent)을 사용하여 구한다



### XGBoost (eXtra Gradient Boost)
- 일반 GBM보다 속도가 빠름, 자체 과적합 규제 기능으로 과적합에 강하다.
- Tree Pruning(나무 가지치기)으로 긍정 이득이 없는 분할을 가지치기해서 분할 수를 줄임
- Early Stopping(조기 중단) 기능으로 오차가 줄지 않을 경우 실제 estimators 갯수만큼 학습하지 않고 학습을 중단
- 원래는 C/C++ 용 라이브러리인데 XGBoost 개발 그룹에서 나중에 파이썬용 사이킷런용 제공

* (1) 파이썬 래퍼 XGBoost : xgboost ( DMatrix 객체 사용, train()/predict() 함수)
* (2) 사이킷런 래퍼 XGBoost : XGBoostClassifier, XGBoostRegressor ( fit()/predict() 함수)

XGBoost : 
- 결손치를 대처하는 알고리즘 , 결손치를 drop 또는 대입값이 필요 없다 
- 결손치가 있는 데이터를 학습시킬 수 있음

    - 결손치와 그렇지 않은 데이터로 분할
    - 깨끗한 데이터 분리하고 결정트리로 분할을 실시한다

#### LightGBM이란?
    - LightGBM (Light Gradient Boosting Machine) 은 Microsoft Research (MSR)를 중심으로 한 팀에서 발표 된 기계 학습 알고리즘
    - XGBoost와 같은 결정 트리의 경사 부스팅 알고리즘 이며, XGBoost에 비해 고속이면서 가벼운특징을 가진다.
    - 2016년에 발표된 이래 Kaggle 등의 다양한 콘페의 상위 모델로 채택되고 있다.
    - XGBoost등의 종래의 결정 나무의 경사 부스팅은 매우 높은 정밀도를 기대하는 반면, 결정 나무를 만들 때의 「분할」에 시간이 걸리는 것이 단점이있다. (연속값의 특징량에 있어서 손실이 최소가 되도록 2분할하려면, 미리 정렬된 데이터에 대해, 어디에서 분할하는 것이 좋은가를 하나하나 계산해 최종적인 분기점을 결정하기 때문에, 데이터의 수만큼 계산량이 증가한다.)
- LightGBM 특징
    - level wise에서 leaf wise: 손실이 가장 줄어드는 노드에서 나누기
    - histogram based: 연속값의 특징량을 히스토그램화하여 bin 단위로 분할
    - Gradient-based One-Side Sampling (GOSS): 기울기가 작은 데이터는 랜덤 샘플링
    - Exclusive Feature Bundling (EFB): 복수의 특징량을 bundle(0이 아닌 데이터)하고 정리해 하나의 특징량과 같이 취급 
    
- 정렬된 값에 대해 하나씩 분기점에서 손실이 어떻게 변화하는지 결정
- 분할시에 계산량은 데이터 수, 특징량의 수로 판단한다
- 분기점의 검색에 걸리는 비용은 bin의 수 n한다

```python
import pandas as pd
import seaborn as sns

df = sns.load_dataset('titanic')
X = df.loc[:, (df.columns!='survived') & (df.columns!='alive')]

X = pd.get_dummies(X, drop_first=True)
y= df['survived']
```

```python
# 데이터 분할
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size=0.3, random_state=0)
```

### LGBMClassifier param

class LGBMClassifier(sklearn.base.ClassifierMixin, LGBMModel)
LGBMClassifier(
1. boosting_type: str = 'gbdt',
 - gbdt’, traditional Gradient Boosting Decision Tree. ‘dart’, Dropouts meet Multiple Additive Regression Trees. ‘goss’, Gradient-based One-Side Sampling. ‘rf’, Random Forest.
2. learning_rate (float, optional (default=0.1)) 학습률
3. num_leaves: int = 31,
4. max_depth: int = -1,
5. earning_rate: float = 0.1,
6. n_estimators: int = 100, 부스팅 횟수 제어
)

```python
print(help(lgb.LGBMClassifier))
```

```python
import lightgbm as lgb # 분류 회귀 다 됨
model = lgb.LGBMClassifier(boosting_type='goss',max_depth=5, random_state=0)
eval_set = [(X_test, y_test)]
callbacks = []  # 리스트 객체 사용자함수를 담을 곳
callbacks.append(lgb.early_stopping(stopping_rounds=10)) # 부스팅 반복 횟수 제어
callbacks.append(lgb.log_evaluation())
model.fit(X_train, y_train, eval_set=eval_set, callbacks=callbacks)
```

```python
print(model.best_iteration_)
print(model.best_score_)
```

```python
from sklearn import metrics
y_pred = model.predict_proba(X_test)
metrics.log_loss(y_test, y_pred)
```

```python
lgb.plot_metric(model) # 학습 곡선 확인 : 학습 데이터와 검증 데이터에 대한 평가
# 그래프가 0으로 내려가지 않는다 -> 학습율을 낮춘다 , 
```

```python
# 특징량 중요도
lgb.plot_importance(model)
```

```python
# XGboost 이용
import xgboost as xgb
import pandas as pd
import seaborn as sns
from xgboost import XGBClassifier

df = sns.load_dataset('titanic')
X = df.loc[:, (df.columns!='survived') & (df.columns!='alive')]

X = pd.get_dummies(X, drop_first=True)
y= df['survived']

# 데이터 분할
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size=0.3, random_state=0)

model = xgb.XGBClassifier(early_stopping_rounds=10)
eval_set = [(X_test, y_test)]
model.fit(X_train, y_train, eval_set=eval_set, verbose=True)
```

```python
print(model.best_iteration)
print(model.best_score) 
# help(model)
```

```python
# 예측

from sklearn import metrics
y_pred = model.predict_proba(X_test)
metrics.log_loss(y_test, y_pred)
```
